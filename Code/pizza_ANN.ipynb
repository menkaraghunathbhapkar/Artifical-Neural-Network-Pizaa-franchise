```python
image_size =50
path1=r"G:\DL Paper 1\Datasets\ANN\Pizza Refund\train"
cate=['Burnt_pizza','Good_pizza']
input_image = []
for i in cate:
    folders = os.path.join(path1 ,i)
    label = cate.index(i)
    for image in os.listdir(folders):
        image_path = os.path.join(folders , image)
        if(image_path[-3:]!="gif"):
            image_array = cv2.imread(image_path,0)
            image_array =cv2.resize(image_array,(image_size,image_size))
            input_image.append([image_array , label])
```

```python
len(input_image)
```

```python
#---------------------------------------------------------
np.random.shuffle(input_image) ## to shuffle the images
#--------------------------------------------------------

X = []
Y = []

for X_values , labels in input_image:
    X.append(X_values.reshape(50*50))
    Y.append(labels)

#--------------------------------------
X = np.array(X)
Y = np.array(Y)

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=.2,random_state=101)
```

# model Building

```python
model=tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(512,activation=tf.nn.relu,input_shape=X[0].shape))

model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))

model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))
model.add(Dropout(0.4))


# output layer

model.add(tf.keras.layers.Dense(2,activation=tf.nn.softmax))

model.compile(optimizer="adam",loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model1=model.fit(x_train,y_train,epochs=20,validation_split=.2,batch_size=100)#epoch is no of ieration of model after passig full data # batch_size=no of records that we want to send each time  
```

```python
model1.history
```

```python
df=pd.DataFrame(model1.history)
```

```python
plt.plot(df.loss,label="loss")
plt.plot(df.val_loss,label="validation_Loss")
plt.legend()
```

```python
plt.plot(df.accuracy,label="loss")
plt.plot(df.val_accuracy,label="validation_Loss")
plt.legend()
```

```python
# predict using model
```

```python
pred=model.predict_classes(x_train)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(pred,y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_train,pred)
```
